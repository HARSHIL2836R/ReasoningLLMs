```python
import matplotlib.pyplot as plt

# Sample data (replace these with your actual values)
n_samples = [1,4,6,8,10,12,20]
accuracy_greedy = [70, 77, 79, 78, 79, 77, 78]
accuracy_sc_with_clustering = [72, 73, 78, 81, 77, 79, 80]

# Plot
plt.figure(figsize=(8, 5))
plt.plot(n_samples, accuracy_greedy, marker='o', color='blue', label='Greedy')
plt.plot(n_samples, accuracy_sc_with_clustering, marker='s', color='green', label='SC with Clustering')

# Labels and title
plt.xlabel("Number of Samples")
plt.ylabel("Accuracy (%)")
plt.xticks(range(0,21,2))
plt.title("Accuracy vs Number of Samples (for temperature = 0.7)")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
